```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyArray = np.linspace(0,10,20)
```

```python
numpyArray
```

```python
numpyArray2 = numpyArray ** 2
```

```python
(myFigure, myEksen) = plt.subplots()
myEksen.plot(numpyArray, numpyArray2,color="#C96F23",alpha = 0.5) #rgb hex code degerleri --- alpha degeri ise transparanlik ekler 1 mat olarak gecer.
myEksen.plot(numpyArray2,numpyArray,color = "#3A95A8")
```

```python
(yeniFigure,yeniEksen)= plt.subplots()
yeniEksen.plot(numpyArray,numpyArray + 2 ,"--",color="blue", linewidth = 5.0) #--> linewidth komutu cizginin kalinligini ayarlar
yeniEksen.plot(numpyArray,numpyArray + 3 ,"--",color="red", linewidth = 2.9)
yeniEksen.plot(numpyArray,numpyArray + 4 ,color="yellow", linewidth = 2.0,linestyle = ":")
yeniEksen.plot(numpyArray,numpyArray + 5 ,color="pink", linewidth = 2.0,linestyle = "-.")
yeniEksen.plot(numpyArray,numpyArray + 6 ,color="#000000", linewidth = 2.0, marker = "o",markerfacecolor = "red", markersize = 5.0)
```
