```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yasListesi = [10,20,30,30,40,50,60,70,70]
```

```python
kiloListesi = [20,60,80,85,86,87,95,90,99]
```

```python
numpyYasList = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasList
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasList,numpyKiloListesi,"g")
```

```python
plt.plot(numpyYasList,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yasa Gore Degisim Grafigi")
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 3
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1,numpyArray2,"g-*")
```

```python
plt.subplot(2,2,1)
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.subplot(2,2,2) # 2 sira olacak 2 kolon olacak su anda 2. grafigi ciziyorum
plt.plot(numpyArray2,numpyArray1,"g--")
plt.subplot(2,2,3)
plt.plot(numpyYasList,numpyKiloListesi,"r*-")
plt.subplot(2,2,4)
plt.plot(numpyKiloListesi,numpyYasList,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])#(GRAFIK ICI X,GRAFIK ICI Y,x ekseni uzunluk,y ekseni uzunluk)
figureAxes.plot(numpyArray1,numpyArray2,"g--")
figureAxes.set_xlabel("x Ekseni")
figureAxes.set_ylabel("y Ekseni")
figureAxes.set_title("Grafik Basligi")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyArray1,numpyArray2,"g--")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Baslik")

eksen2.plot(numpyArray2,numpyArray1,"b--")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Kucuk Grafik Baslik")

```

```python
plt.subplot()
```

```python
(myFigure, myEksen) = plt.subplots(nrows=1,ncols=2)
type(myEksen)

for eksen in myEksen:
  eksen.plot(numpyArray1,numpyArray2,"g--")
  eksen.set_xlabel("X Label")
  eksen.set_ylabel("Y Label")

plt.tight_layout() #Tablolari birbirinden ayirir ve karmasikligi onler duzenler.!!!!
```

# GÖRSELLEŞTİRME


```python
newFigure = plt.figure()
#figsize=(6,6)
#dpi = 150

yeniEksen = newFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyArray1,numpyArray1**1.2,"g--",label = "numpyDizisi ** 1.2")
yeniEksen.plot(numpyArray1,numpyArray1**1.4,"b*-",label = "numpyDizisi ** 1.4")
yeniEksen.plot(numpyArray1,numpyArray1**1.6,"r",label = "numpyDizisi ** 1.6")
yeniEksen.legend(loc=0)
```
